```python
import numpy as np
import yfinance as yf
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd
import random
from itertools import zip_longest
```

```python
START = "2021-01-01"
END = "2025-01-01"
TICKER = "^GSPC"
TITLE = TICKER + " - " +(datetime.now()).strftime("%Y-%m-%d - %H:%M:%S")
```

```python
data = pd.read_csv("data.csv",index_col=0)

# data = yf.download(TICKER,start=START, end=END,multi_level_index=False)

data["log_return"] = np.log(data["Close"] / data["Close"].shift(1))

data.dropna(inplace=True)
```

```python
fig = plt.figure(figsize=(15,4))
sns.lineplot(data["log_return"].cumsum()).set_title(TITLE)
plt.show()
```

```python
blocks = []

for x in range(len(blocks_index)):
    if x == 0:
        continue
    blocks.append((int(blocks_index[x-1]),int(blocks_index[x])))

print(blocks)
```

> To generate the MBB samples, we select $b$ blocks at random with replacement from the collection $\{B_1, . . . , B_N\}$.

```python
boot = []
for _ in range(100):
    boot.append(random.choices(blocks,k=len(blocks)))

# print(boot)
```

```python
fig, ax = plt.subplots(figsize=(15,4))
boot_data = [np.concat([data["log_return"].values[block[x][0] : block[x][1]] for x in range(len(block))]) for block in boot]

boot_df = pd.DataFrame(zip_longest(*boot_data, fillvalue=np.nan),columns=[f"log_returns_{x}" for x in range(len(boot_data))])

sns.lineplot(boot_df.cumsum(),dashes=False,legend=False).set_title("Block bootstrapping trial on S&P500 - block size: 10")

ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
```

```python
def create_sma(data,sma1=20,sma2=50):
    data = data[:]
    price = np.array([100]*len(data))
    price = price * np.exp(data.cumsum())
    return pd.DataFrame({"price":price,"log_return":data,"short":price.rolling(sma1).mean(), "long":price.rolling(sma2).mean()})
    

create_sma(boot_df["log_returns_0"])
```

```python
sim_data = create_sma(boot_df["log_returns_5"])

sns.lineplot(sim_data[["price","short","long"]])
```

```python
equity_data = pd.DataFrame([],columns=boot_df.columns,index=boot_df.index) 
trades = []

for boot_sim in boot_df:
    
    sim_data = create_sma(boot_df[boot_sim])
    
    equity_data[boot_sim] = equity_data[boot_sim].where(sim_data["short"] > sim_data["long"], 1) * sim_data["price"]
    trades.append(equity_data[boot_sim].where(sim_data["short"] > sim_data["long"], 1).sum())
    # equity_data[boot_sim] = equity_data[boot_sim].where(sim_data["short"] < sim_data["long"],-1)

equity_data = equity_data.ffill()
```
